```python
df.sort_values(by="rating", ascending=False)
```

```python
import matplotlib.pyplot as plt # henüz o kadar ilerlemesem de bir grafik yapmak istiyorum
```

```python
# np.round(rating.values).astype(int)
plt.pie(rating.values, labels= rating.index)
plt.show()
# üreticilerin ne kadar ürünü var tablosu
```
