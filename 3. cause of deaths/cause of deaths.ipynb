```python
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv("cause_of_deaths.csv")
df = dataset.copy()
df.head()
```

```python
df.shape, df.columns, df.dtypes
```

```python
df["Country/Territory"].unique()
```

```python
turkiye = df.loc[df["Code"] == "TUR"]
turkiye.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 6))
sns.lineplot(x = "Year", y = "Conflict and Terrorism", data = df);
# number of death by conflich and terrorism by years in turkey
```

```python
turkiye.iloc[:, 3:].sum(axis=1).values
```

```python
plt.figure(figsize=(16, 6))
sns.lineplot(x = turkiye["Year"],
             y = turkiye.iloc[:, 3:].sum(axis=1).values
            );
# number of all deaths by years in turkey 
```

```python
tr_2019 = turkiye.loc[df.Year == 2019]
tr_2019
```

```python
# deaths in turkey in 2019 transferred to pie charts
plt.figure(figsize=(25, 9))
plt.pie(tr_2019.iloc[:, 3:].values[0],
    labels = tr_2019.columns[3:]
)
plt.show()
```
